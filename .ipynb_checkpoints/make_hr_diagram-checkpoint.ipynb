# make_hr_diagram Jupyter Notebook

This notebook reads in the Python module hr_diagram and generates a HR diagram from an ASCII data file read in. The HR diagram is also output to a PDF file. 

The sytax to be used is:
<code>
hr_diagram.generate_plot (inputfile, outputfile)
</code>

where inputfile and outputfile are two file strings. So, for example: 
<code>
hr_diagram.generate_plot ("stars_dist_limited.dat", "stars_dist_limited2.pdf")
</code>


```python
#!/usr/bin/python

import matplotlib.pyplot as plt

import hr_diagram

%pylab inline
pylab.rcParams['figure.figsize'] = (20, 12)
```

```python
hr_diagram.generate_plot ("stars_dist_limited.dat", "stars_dist_limited.pdf")
```
