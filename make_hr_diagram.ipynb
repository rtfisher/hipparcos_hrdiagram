# make_hr_diagram Jupyter Notebook

This notebook reads in the Python module hr_diagram and generates a HR diagram from an ASCII data file read in.  

The sytax to be used is:
<code>
hr_diagram.generate_plot (inputfile)
</code>

where inputfile is a string with the name of the file. So, for example: 
<code>
hr_diagram.generate_plot ("stars_dist_limited.dat")
</code>


```python
#!/usr/bin/python

import matplotlib.pyplot as plt

import hr_diagram

%pylab inline
pylab.rcParams['figure.figsize'] = (20, 12)
```
